```python
import pandas as pd
import numpy as np

import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as pl

matplotlib.rc('text', usetex=True)
pl.rcParams['text.latex.preamble'] = [
    r'\usepackage{tgheros}',    
    r'\usepackage[eulergreek]{sansmath}',   
    r'\sansmath',
    r'\usepackage{siunitx}',    
    r'\sisetup{detect-all}'
]

# library to generate the map
import folium
import folium.plugins as plugins

# generates the legend 
from utils.legend import template_macro

# custom date-time util
from utils.date_time import yyyy_mm_dd

# distance between two coordinate points
from utils.lonlat_distance import coordinate_distance

import itertools
```

```python
locs = pd.read_csv('data/static_locs.csv', sep=';', 
                   dtype={'latitude': np.float64, 'longitude': np.float64})
sens = pd.read_csv('data/sensor_locs.csv', sep=';', 
                   dtype={'latitude': np.float64, 'longitude': np.float64})

nox_all = pd.read_csv('data/NOX_Juli-Sept.csv', sep=',', decimal=",")

pc_vals = pd.read_csv('data/pc_vals.csv', sep=',')

pred = {'180704': np.load('data/regression/2018-07-04_prediction.npy'),
        '180708': np.load('data/regression/2018-07-08_prediction.npy')}

pred_nT = {'180704_no-block': np.load('data/regression/2018-07-04_prediction_notInTrain_no-block.npy'),
           '180708_no-block': np.load('data/regression/2018-07-08_prediction_notInTrain_no-block.npy'),
           '180704_small-block': np.load('data/regression/2018-07-04_prediction_notInTrain_small-block.npy'),
           '180708_small-block': np.load('data/regression/2018-07-08_prediction_notInTrain_small-block.npy'),
           '180704_medium-block': np.load('data/regression/2018-07-04_prediction_notInTrain_medium-block.npy'),
           '180708_medium-block': np.load('data/regression/2018-07-08_prediction_notInTrain_medium-block.npy'),
           '180704_large-block': np.load('data/regression/2018-07-04_prediction_notInTrain_large-block.npy'),
           '180708_large-block': np.load('data/regression/2018-07-08_prediction_notInTrain_large-block.npy')}
```

```python
# calculate nodes 


def extract_locs(stem_node, distance_excl):

    lat, lon = locs[locs['ElemUID']==stem_node]['latitude'], locs[locs['ElemUID']==stem_node]['longitude']

    excl = []

    for index, row in locs.iterrows():
        x = coordinate_distance(row['latitude'], row['longitude'], lat, lon)
        if x < distance_excl:
            excl.append(int(row['ElemUID']))
            
    return excl
            
            

stem_node, distance_excl = 18364005, 14
excl_set1 = extract_locs(stem_node, distance_excl)       
excl_set1.extend([18360002, 18360001, 18361004, 18361003, 18361002, 18361006, 18361005, 18361001]) 


stem_node, distance_excl = 18366003, 22
excl_set2 = extract_locs(stem_node, distance_excl)
excl_set2.extend(excl_set1)
excl_set2 = list(np.unique(excl_set2))

stem_node, distance_excl = 18887020, 24
excl_set3 = extract_locs(stem_node, distance_excl)   
excl_set3.extend(excl_set2)
excl_set3 = list(np.unique(excl_set3))

excl_set0 = []

# generally can test for difference sizes of a road-block 

roadblock_sets = [excl_set0, excl_set1, excl_set2, excl_set3]
labels = ['no-block', 'small-block', 'medium-block', 'large-block']
```

```python
days, weekday = ['180704', '180708'], ['Mittwoch','Sonntag']
time_pts = [7,13,16,19,24,29,34,39,44]
#time_pts = [19]

for (wd,day),(exnodes,label) in itertools.product(zip(weekday,days), zip(roadblock_sets, labels)):
    
    '''main loop; generate maps for every time point on every day'''
    
    # load traffic data    
    df = pd.read_csv('data/30m/FFM_DEZ_'+day+'_30min.csv', 
                     error_bad_lines=False, encoding='ISO-8859-1', 
                     sep=',')
    
    # filter nox values for day in question
    # convert 180703 to 2018-07-03
    noxd = nox_all[(nox_all['time'] > yyyy_mm_dd(day)+' 00:00') \
                   & (nox_all['time'] < yyyy_mm_dd(day)+' 23:59')]
    
    
    for tp in time_pts:
        
        # -------------- create nox curve graphic ---------------
        
        fname = day+'_%.2d'%(tp)
        
        fig, ax = pl.subplots()
        fig.set_size_inches(2,1)
        ax.plot(pred_nT[day+'_'+label], color='grey', lw=2)
        ax.plot(list(noxd['NOX']), color='#E37222', lw=2);
        
        #print(list(noxd['time'])[tp])
        pl.xticks([7,23,39], labels=['4:00','12:00','20:00'])

        ax.spines['right'].set_visible(False)
        ax.spines['top'].set_visible(False)
        ax.yaxis.set_ticks_position('left')
        ax.xaxis.set_ticks_position('bottom')

        ax.annotate('',xy=(tp,list(noxd['NOX'])[tp]), 
                    xytext=(tp,list(noxd['NOX'])[tp]+40),
                    arrowprops=dict(facecolor='black', 
                                    shrink=0.2, headwidth=3, 
                                    headlength=5, width=1))

        ax.set_ylim(0,150)
            
        fig.savefig('maps/'+label+'/gfx/'+fname+'.png', dpi=200, bbox_inches='tight')
        
        
        # ---------------- create folium map --------------------        
        
        folium_map = folium.Map(location=[locs['latitude'][0], locs['longitude'][0]],
                            zoom_start=14,tiles='cartodbpositron')
                            #tiles="CartoDB dark_matter")
        
        fail = 0

        for index, row in locs.iterrows():

            try:
                # nodes with radius zero get displayed as relatively large circles
                # and shouldn't get displayed at all
                if int(df.iloc[tp][str(int(row['ElemUID']))]) > 0:

                    popup_text = """<br>
                                Spulen-ID: {}<br>
                                Fahrten: {}"""
                    popup_text = popup_text.format(int(row['ElemUID']),
                                                   int(df.iloc[tp][str(int(row['ElemUID']))]))
                                     
                    if row['ElemUID'] in exnodes:

                        marker = folium.CircleMarker(location=[row['latitude'], row['longitude']], 
                                                     radius=3,
                                                     popup=popup_text, fill=True, color='red')
                        marker.add_to(folium_map)
                        
                    else:
                        
                        marker = folium.CircleMarker(location=[row['latitude'], row['longitude']], 
                                                     radius=int(df.iloc[tp][str(int(row['ElemUID']))])/25.,
                                                     popup=popup_text, fill=True)
                        marker.add_to(folium_map)

            except KeyError:
                fail+=1


        print('Could not add ', fail, 'ElemUIDs')  

        for index, row in sens.iterrows():
            if index==1:
                marker = folium.CircleMarker(location=[row['latitude'], row['longitude']], 
                                             color="#E37222", fill=True, 
                                             radius=noxd.iloc[tp]['NOX']/2)
                marker.add_to(folium_map) 


        print(noxd.iloc[tp]['NOX'])
        
        macro = template_macro(wd+' '+day[5]+'.'+day[3]+'.'+'2018')           
        folium_map.get_root().add_child(macro)

        url = 'gfx/'+fname+'.png'

        plugins.FloatImage(url, bottom=2.5, left=2.5).add_to(folium_map)

        #folium.TileLayer('stamentoner').add_to(folium_map)

        folium_map.save('maps/'+label+'/'+fname+'.html')

        folium_map
```
